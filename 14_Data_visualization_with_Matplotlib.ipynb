#Line plot


```python
# Creating a DataFrame
data = {'Year': [2018, 2019, 2020, 2021, 2022],
        'Sales': [250, 300, 400, 350, 500]}
df_1 = pd.DataFrame(data)

# Generating Line Plot
plt.plot(df_1['Year'], df_1['Sales'])

# Adding title and labels
plt.title('Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')

# Display the plot
plt.show()
```

#Bar plot

```python
# Creating a DataFrame
data = {'Category': ['Coca Cola', 'Sprite', 'Fanta', 'Ciel'],
        'Values': [500, 100, 150, 400]}
df_2 = pd.DataFrame(data)

# Generating Bar Plot
plt.bar(df_2['Category'], df_2['Values'])

# Adding title and labels
plt.title('Category Comparison')
plt.xlabel('Category')
plt.ylabel('Value')

# Display the plot
plt.show()
```

```python
# Path to the CSV file on Google Drive
# Remember to replace 'my_folder' with the name of the folder where you uploaded the file
csv_path = '/content/drive/MyDrive/my_folder/scores.csv'

# Read the CSV file
df_3 = pd.read_csv(csv_path)

# Generating the histogram
plt.hist(df_3['Scores'], bins=10, color='green', edgecolor='black')

# Adding title and labels
plt.title('Distribution of Product Scores')
plt.xlabel('Scores')
plt.ylabel('Number of Clients')

# Adding a grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

#Scatter Plot

```python
# Path to the CSV file on Google Drive
# Remember to replace 'my_folder' with the name of the folder where you uploaded the file
csv_path_mkt = '/content/drive/MyDrive/my_folder/sales_mkt.csv'

# Read the CSV file
df_4 = pd.read_csv(csv_path_mkt)

# Generating the Scatter Plot
plt.scatter(df_4['Marketing Spend'], df_4['Sales'], color='blue', alpha=0.6, edgecolor='black')

# Adding title and labels
plt.title('Scatter Plot: Marketing Spend vs Sales')
plt.xlabel('Marketing Spend ($)')
plt.ylabel('Sales (Units)')

# Display the plot
plt.show()
```

# Pie Chart

```python
# Creating a DataFrame
data = {'Product': ['Fanta','Ciel', 'Coca Cola','Sprite' ],
        'Percentage': [15, 30, 45, 10]}
df_5 = pd.DataFrame(data)

# Generating Pie Chart
plt.pie(df_5['Percentage'], labels=df_5['Product'], autopct='%1.1f%%')

# Adding title
plt.title('Sales Proportion')

# Display the plot
plt.show()
```

# Area Plot


```python
# Path to the CSV file on Google Drive
# Remember to replace 'my_folder' with the name of the folder where you uploaded the file
csv_path_month = '/content/drive/MyDrive/my_folder/monthly_sales.csv'

# Read the CSV file
df_6 = pd.read_csv(csv_path_month)

# Generating the area plot
df_6.plot(x='Month', y=['Product Line A', 'Product Line B'], kind='area', stacked=True, alpha=0.6)

# Adding title and labels
plt.title('Monthly Sales of Product Lines A and B')
plt.xlabel('Month')
plt.ylabel('Sales (in Thousands)')

# Display the plot
plt.show()
```

#Subplots

```python
# Path to the CSV file on Google Drive
# Remember to replace 'my_folder' with the name of the folder where you uploaded the file
csv_path_retail = '/content/drive/MyDrive/my_folder/retail_data.csv'

# Read the CSV file
df_7 = pd.read_csv(csv_path_retail)

# Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Subplot 1: Sales over time
axs[0, 0].plot(df_7['Month'], df_7['Sales'], marker='o')
axs[0, 0].set_title('Monthly Sales')
axs[0, 0].set_xlabel('Month')
axs[0, 0].set_ylabel('Sales (in Thousands)')

# Subplot 2: Customer Traffic over time
axs[0, 1].bar(df_7['Month'], df_7['Customer Traffic'], color='orange')
axs[0, 1].set_title('Customer Traffic')
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Number of Customers')

# Subplot 3: Average Transaction Value over time
axs[1, 0].plot(df_7['Month'], df_7['Avg Transaction Value'], color='green', marker='s')
axs[1, 0].set_title('Average Transaction Value')
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Value (in Dollars)')

# Subplot 4: Return Rate over time
axs[1, 1].bar(df_7['Month'], df_7['Return Rate (%)'], color='red')
axs[1, 1].set_title('Return Rate')
axs[1, 1].set_xlabel('Month')
axs[1, 1].set_ylabel('Return Rate (%)')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Path to the CSV file on Google Drive
# Remember to replace 'my_folder' with the name of the folder where you uploaded the file
csv_path_retail = '/content/drive/MyDrive/my_folder/retail_data.csv'

# Read the CSV file
df_7 = pd.read_csv(csv_path_retail)

# Creating subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Subplot 1: Sales over time
axs[0, 0].plot(df_7['Month'], df_7['Sales'], marker='o')
axs[0, 0].set_title('Monthly Sales')
axs[0, 0].set_xlabel('Month')
axs[0, 0].set_ylabel('Sales (in Thousands)')

# Subplot 2: Customer Traffic over time
axs[0, 1].bar(df_7['Month'], df_7['Customer Traffic'], color='orange')
axs[0, 1].set_title('Customer Traffic')
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Number of Customers')

# Subplot 3: Average Transaction Value over time
axs[1, 0].plot(df_7['Month'], df_7['Avg Transaction Value'], color='green', marker='s')
axs[1, 0].set_title('Average Transaction Value')
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Value (in Dollars)')

# Subplot 4: Return Rate over time
axs[1, 1].bar(df_7['Month'], df_7['Return Rate (%)'], color='red')
axs[1, 1].set_title('Return Rate')
axs[1, 1].set_xlabel('Month')
axs[1, 1].set_ylabel('Return Rate (%)')

# Exporting the figure
plt.savefig('retail_subplot.png', dpi=300)

# Adjust layout
plt.tight_layout()
plt.show()
```
